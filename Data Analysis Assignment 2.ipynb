```python
#3.) You will see that Sentiment can take 5 values: "Extremely Negative", "Negative", "Neutral", "Positive", "Extremely Positive". 
#Using an appropriate color palette and an appropriate type of graph, visualise how Sentiment is distributed.
df.dropna(subset = "Sentiment", inplace=True)
for i in range(df.Sentiment.values.size): 
    df.Sentiment.values[i] = re.sub("^Extremely Negative", "aExtremely Negative", df.Sentiment.values[i])
    df.Sentiment.values[i] = re.sub("^Negative", "bNegative", df.Sentiment.values[i])
    df.Sentiment.values[i] = re.sub("^Neutral" , "cNeutral", df.Sentiment.values[i])
    df.Sentiment.values[i] = re.sub("^Positive" , "dPositive", df.Sentiment.values[i])
    df.Sentiment.values[i] = re.sub("^Extremely Positive" , "eExtremely Positive", df.Sentiment.values[i])
df_sent = df.groupby(by = "Sentiment")
x = range(5)
z = ["Ex. Negative", "Negative", "Neutral", "Positive", "Ex. Positive"]
colours = ["#ff3333", "#960000", "gray", "#4e9117", "#80FF00"]
plt.bar(x, df_sent["OriginalTweet"].count(), edgecolor="black", color = colours)
plt.xticks(x, z)
plt.title("No. Tweets by Sentiment")
plt.ylabel("No. Tweets")
plt.xlabel("Sentiment")
plt.show()
for i in range(df.Sentiment.values.size): 
    df.Sentiment.values[i] = re.sub("^a", "", df.Sentiment.values[i])
    df.Sentiment.values[i] = re.sub("^b", "", df.Sentiment.values[i])
    df.Sentiment.values[i] = re.sub("^c" , "", df.Sentiment.values[i])
    df.Sentiment.values[i] = re.sub("^d" , "", df.Sentiment.values[i])
    df.Sentiment.values[i] = re.sub("^e" , "", df.Sentiment.values[i])
```

```python
#4.) Change the labels of part of the data.
#Combine "Extremely Negative" and "Negative" into “Negative” and "Positive" and "Extremely Positive" into positive.
for i in range(df.Sentiment.values.size): 
    df.Sentiment.values[i] = re.sub("^Extremely ", "", df.Sentiment.values[i])
df_sent = df.groupby(by = "Sentiment")
x = range(3)
z = ["Negative", "Neutral", "Positive"]
colours = ["#ff3333", "gray", "#80FF00"]
plt.bar(x, df_sent["OriginalTweet"].count(), edgecolor="black", color = colours)
plt.xticks(x, z)
plt.title("No. Tweets by Sentiment")
plt.ylabel("No. Tweets")
plt.xlabel("Sentiment")
plt.show()
```

```python
#PART 3
#Perform the following text cleaning steps.
for i in range(df.OriginalTweet.values.size):
    #1.) Remove hyperlinks.
    df.OriginalTweet.values[i] = re.sub("https?:\/\/[a-z0-9-]+(?:\.[a-z0-9-]+)+[/?#]\S*", "", df.OriginalTweet.values[i])

    #2.) Remove hashtags.
    df.OriginalTweet.values[i] = re.sub("[#]", "", df.OriginalTweet.values[i])

    #3.) A.) Replace \n and \r with an empty space.
    df.OriginalTweet.values[i] = re.sub(r"\n", " ", df.OriginalTweet.values[i])
    df.OriginalTweet.values[i] = re.sub(r"\r", " ", df.OriginalTweet.values[i])

    #3.) B.) Convert to lowercase.
    df.OriginalTweet.values[i] = df.OriginalTweet.values[i].lower()
                                        
```

```python
#4.) Using TweetTokenizer in nltk.tokenize, perform tokenization
#5.) In the same function, remove stopwords and punctuation signs
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import TweetTokenizer
stop_words = set(stopwords.words("english"))
tk = TweetTokenizer()

def tokenClean(tweet):

    tweet = re.sub("[!#$%&'()*+,-./:;<=>?@[\]^_`{|}~]", "", tweet)
    tweet = tk.tokenize(tweet)
    tweet = [word for word in tweet if word not in stop_words]
    return tweet

df.dropna(subset="OriginalTweet", inplace=True)

df["cleanedTweets"] = df["OriginalTweet"].apply(tokenClean)
```

```python
#6. Perform stemming using PorterStemmer in nltk.stem
from nltk.stem import PorterStemmer

ps = PorterStemmer()

for tweet in range(df["cleanedTweets"].size):

    for word in range(len(df["cleanedTweets"].iloc[tweet])):

            df["cleanedTweets"].iloc[tweet][word] = ps.stem(df["cleanedTweets"].iloc[tweet][word])
```

```python
#7.) Apply these functions to the dataset and visualise the first 5 rows. 
from wordcloud import WordCloud
text = ""
for tweet in range(5):
    text += " ".join(word for word in df["cleanedTweets"].iloc[tweet])
wordcloud = WordCloud().generate(text)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
display(df["cleanedTweets"].head())
plt.show()
```

```python

#PART 4
#1.) Create a tf-idf vectorizer and fit the function.
from sklearn.feature_extraction.text import TfidfVectorizer

arr = [None] * df["cleanedTweets"].size

text = ""

for x in range(df["cleanedTweets"].size):
    
    text = " ".join(tweet for tweet in df["cleanedTweets"].iloc[x])
    arr[x] = text

tfidf = TfidfVectorizer()

sparse_matrix = tfidf.fit_transform(arr)
```

```python
#2.) Keeping only the columns Tweet and Sentiment, split your dataset into training (80%) and test (20%) sets.

df = df.drop(columns = ["Location", "OriginalTweet", "UserName", "ScreenName", "TweetAt"])

X = sparse_matrix
y = df["Sentiment"]

from sklearn.model_selection import train_test_split
# Split the data into training, validation, and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42, shuffle=True)
print(f"Training set size: {X_train.shape[0]}")
print(f"Test set size: {X_test.shape[0]}")
```

```python
#3.) Build a logistic regression model and fit the data.
from sklearn import linear_model
from sklearn.metrics import accuracy_score
logisticR = linear_model.LogisticRegression(max_iter = 1000)
logisticR.fit(X_train, y_train)
```

```python
#4.) What is the accuracy of the model?

print(str(accuracy_score(y_test, logisticR.predict(X_test))) + "% accuracy.")
```

```python
#5.) Generate a classification report
from sklearn.metrics import classification_report
 
report = classification_report(y_test, logisticR.predict(X_test))

print(report)
```
